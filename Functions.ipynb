Q 2. Explain the concept of function arguments and parameters in Python.
- n Python, parameters are variables defined within the parentheses of a function definition, serving as placeholders for values the function expects to receive. Arguments are the actual values passed to the function when it is called.
For instance, in the function definition def add(x, y):, x and y are parameters. When calling the function with add(5, 3), 5 and 3 are the arguments.
Python supports various types of arguments:

Positional arguments: Passed in the order they are defined in the function.

Keyword arguments: Passed with the parameter name, allowing for
order-independent passing.

Default arguments: Parameters with default values, used if no argument is provided in the call.

Variable-length arguments: Allow passing a variable number of arguments using *args (for positional) or **kwargs (for keyword).

Q 3. What are the different ways to define and call a function in Python?
- In Python, functions can be defined and called in several ways:
1. Standard Function Definition
This is the most common way to define a function, using the def keyword, followed by the function name, parentheses for parameters, and a colon. The function body is indented below.


```python
def greet(name):
    return f"Hello, {name}!"

# Calling the function
message = greet("Alice")
print(message)
```

2. Lambda Functions

Lambda functions are small, anonymous functions defined using the lambda keyword. They can take any number of arguments but can only have one expression.

```python
add = lambda x, y: x + y

# Calling the lambda function
result = add(5, 3)
print(result)
```


3. Functions with Default Arguments:

Functions can have default values for arguments, which are used if the argument is not provided in the function call.

```python
def power(base, exponent=2):
    return base ** exponent

# Calling the function with default exponent
square = power(5)
print(square)

# Calling the function with a specified exponent
cube = power(5, 3)
print(cube)
```

4. Functions with Variable Arguments (*args and **kwargs)

*args allows a function to accept any number of positional arguments.

**kwargs allows a function to accept any number of keyword arguments.

```python
def print_args(*args):
    for arg in args:
        print(arg)

print_args(1, "hello", 3.14)


def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(name="John", age=30, city="New York")
```

5. Calling Functions

Functions are called by writing the function name followed by parentheses. If the function takes arguments, they are passed inside the parentheses.

```python
def multiply(x, y):
    return x * y

# Calling the function with positional arguments
product = multiply(4, 7)
print(product)

# Calling the function with keyword arguments
product2 = multiply(x=4, y=7)
print(product2)
```

Q 4. What is the purpose of the `return` statement in a Python function?
- The return statement in Python serves to terminate the execution of a function and optionally send a value back to the caller. When a return statement is encountered, the function's execution halts immediately, and control is passed back to the point where the function was called. If a value or expression follows the return keyword, that value is returned to the caller. Otherwise, if no value is specified, or if there is no return statement at all, the function returns None by default.

```python
def add(x, y):
    return x + y

result = add(5, 3)  # result will be 8
print(result)

def greet(name):
    print("Hello, " + name + "!")
    # No return statement, so it implicitly returns None

return_value = greet("Alice") # Prints "Hello, Alice!"
print(return_value) # return_value will be None
```

Q 6. Explain the concept of generators in Python and how they are defined
- Python generator functions are a powerful tool for creating iterators. In this article, we will discuss how the generator function works in Python.
- A generator function is a special type of function that returns an iterator object. Instead of using return to send back a single value, generator functions use yield to produce a series of results over time. This allows the function to generate values and pause its execution after each yield, maintaining its state between iterations.

Basic Code Example:

```python
def fun(max):
    cnt = 1
    while cnt <= max:
        yield cnt
        cnt += 1

ctr = fun(5)
for n in ctr:
    print(n)
```

Q 9. Explain the purpose and usage of the `map()` function in Python.
- The map() function in Python serves to apply a given function to each item of an iterable (like a list or tuple) and returns an iterator that yields the results. It provides a way to transform or process all elements in an iterable without writing explicit loops.
Python

```python
def square(n):
    return n * n

numbers = [1, 2, 3, 4, 5]

squared_numbers = map(square, numbers)

print(list(squared_numbers))
# Expected Output: [1, 4, 9, 16, 25]
```

Q 10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- map(), filter(), and reduce() are built-in functions in Python that facilitate functional programming paradigms. They operate on iterables (like lists, tuples) and apply a function to the items within them. However, they differ in their purpose and how they transform the data.

map(): Applies a given function to each item in an iterable and returns a new iterator with the results. It transforms each item individually without considering other items.

```python
    numbers = [1, 2, 3, 4]
    squared_numbers = map(lambda x: x**2, numbers)
    print(list(squared_numbers))  # Output: [1, 4, 9, 16]
```

filter(): Creates a new iterator containing only the items from the original iterable for which a given function returns True. It selectively keeps or discards items based on a condition.


```python
    numbers = [1, 2, 3, 4, 5, 6]
    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    print(list(even_numbers))  # Output: [2, 4, 6]
```

reduce(): Applies a function cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value. It combines items together, requiring the functools module in Python 3.

```python
    from functools import reduce
    numbers = [1, 2, 3, 4]
    product = reduce(lambda x, y: x * y, numbers)
    print(product)  # Output: 24
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.
--

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

# Example usage:
my_list = [1, 2, 3, 4, 5, 6]
print(sum_of_even_numbers(my_list))  # Output will be 12 (2 + 4 + 6)
```

Q 4.Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # Check up to the square root of n
        if n % i == 0:
            return False  # Found a factor, not prime
    return True  # No factors found, it is prime

# Testing numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number.")
```

Q 5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

# Example usage
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

Q 6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Example usage
for power in powers_of_two(5):
    print(power)
```

Q 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # Strip removes the trailing newline characters
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found. Please check the path.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
file_path = 'example.txt'  # Replace with your actual file path
for line in read_file_line_by_line(file_path):
    print(line)
```

Q 9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius_temperatures = [0, 20, 30, 40, 100]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Using map() to apply the conversion function to each item in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Output the result
print(fahrenheit_temperatures)
```

Q 10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
# Function to check if a character is not a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Given string
input_string = "Hello, World!"

# Using filter() to keep only non-vowel characters
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Output the result
print(filtered_string)
```
